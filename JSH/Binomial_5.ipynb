El estudiante cree que un buen ensayo necesita recibir y utilizar por lo menos la mitad de los libros pedidos para uno u otro tema seleccionado. Si la probabilidad de que un libro pedido mediante prstamo interbiblioteca llegue a tiempo es de .9 y los libros llegan independientemente uno de otro

      i) Qu tema deber seleccionar el estudiante para incrementar al mximo la probabilidad de escribir un buen ensayo? (Rta: 0.0081)

      ii) Qu pasa si la probabilidad de que lleguen los libros es de slo 0.5 en lugar de 0.9? (Tema A: 0.99; Tema B: 0.9963)
      
      iii) Ninguna tendr que olvidarse.(Tema A: 0.75; Tema B: 0.6875)
