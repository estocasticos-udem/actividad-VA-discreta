2. El nmero promedio de accidentes de trnsito en cierto crucero de carretera es dos por semana. Suponga que el nmero de accidentes sigue una distribucin de Poisson con 

      i) Encuentre la probabilidad de que no haya accidentes en este crucero de carretera durante un periodo de 1 semana. (Rta: 0.135335)

      ii) Encuentre la probabilidad de que haya tres accidentes como mximo en esta seccin de carretera durante un periodo de 2 semanas. (Rta: 0.433471)

      
      

```python
import math

# Tasa de lambda i
lam = 2

# Probabilidad de que no haya accidentes en una semana
prob = math.exp(-lam)
print("La probabilidad de que no haya accidentes en una semana es:", prob)

# Tasa de lambda ii
lam2 = 4

# Probabilidad de que haya tres accidentes como mximo en 2 semanas
prob2 = sum([math.exp(-lam2)*(lam2**k)/math.factorial(k) for k in range(4)])
print("La probabilidad de que haya tres accidentes como mximo en 2 semanas es:", prob2)



```
