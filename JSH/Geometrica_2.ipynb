### Enunciado ###

2. Un contador público certificado (CPA, por sus siglas en inglés) ha encontrado que nueve de cada diez compañías auditadas contienen errores importantes. Si el CPA hace auditoría a una serie de cuentas de empresas

      i) ¿Cuál es la probabilidad de que la primera cuenta que contenga errores importantes sea la tercera en ser auditada?. (Rta: 0.009)

      ii) ¿Cuál es la probabilidad de que la primera cuenta que contenga errores importantes cualquier cuenta después de la segunda? (Rta: 0.01)

        


```python
from scipy.stats import geom

p = 0.9

# Problema 1
k1 = 3
prob1 = (1 - p)**(k1-1) * p
print(f"La probabilidad de encontrar el primer error importante en la tercera cuenta es: {prob1:.3f}")

prob2 = geom.cdf(k1,p) - geom.cdf(2, p)
print(f"La probabilidad de encontrar el primer error importante después de la segunda cuenta es: {prob2:.2f}")

```
